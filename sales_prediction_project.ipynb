item_weight have 2439 null values and outlet size also have 4016 null values . 5681 null values of item_outlet_sales is contributed by test_data , no need to treat those values , since we are not using that data set for model preparation

```python
#treating _null values of item_weight by adding avg value that corresponds to each product
import numpy as np
avg_weight=df.pivot_table(values='Item_Weight',columns='Item_Identifier',aggfunc=np.mean)
avg_weight
```

```python
l=[]
for i in df.index:
  if pd.isnull(df['Item_Weight'][i]):
    l.append(df['Item_Identifier'][i])
    
l=set(l)
l
  
```

```python
for i in l:
  df.fillna(avg_weight[i],inplace=True)
```

```python

df.isnull().sum()
```

```python
#treating _null values of Outlet_Size by adding avg value that corresponds to each product
size_avg=df.pivot_table(values='Outlet_Size',columns='Outlet_Type',aggfunc=lambda x:x.mode())
```

```python
j=[]
for i in df.index:
  if pd.isnull(df['Outlet_Size'][i]):
    j.append(df['Outlet_Type'][i])
j=set(j)
j
   
    
```

```python
for i in j:
  df.fillna(size_avg[i],inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Item_Visibility"].value_counts()
```

879 items have visibility 0 , these values are equal to null .hence we replaced values with mean of item visibility of corresponding products for better performance of the model

```python
item_visibility_avg=df.pivot_table(values='Item_Visibility',columns='Item_Identifier',aggfunc=np.mean)
item_visibility_avg
```

```python
m=[]
for i in df.index:
  if df["Item_Visibility"][i]==0:
    m.append(df['Item_Identifier'][i])
    
m=set(m)
m
  
```

```python
for i in m:
  for j in df.index:
    if df['Item_Identifier'][j] == i:
      df['Item_Visibility'][j]=item_visibility_avg[i]
```

```python
x=train.drop(['Item_Outlet_Sales'],axis=1)
y=train['Item_Outlet_Sales']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

## Linear Regression

```python

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
print(lr.score(x_test,y_test))
```

```python
print(pred)
```

```python
from sklearn import metrics
print('Mean Absolute Error: ',metrics.mean_absolute_error(y_test,pred))
print('Mean Squared Error: ',metrics.mean_squared_error(y_test,pred))
print('Root Mean Squared Error: ',np.sqrt(metrics.mean_squared_error(y_test,pred)))
print('Explained Variance Score: ',metrics.explained_variance_score(y_test,pred))
```
